# Sequence autocorrelation

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
# read match1_momentum
with open('match1_momentum.csv','r') as f:
    reader = csv.reader(f)
    data = list(reader)
data_array = np.array(data)
data_array = data_array[1:,1:]
data_array = data_array.astype(np.float64)
plt.plot(range(300),data_array[:,0])
```

```python
time_series = data_array[:,0]
# Calculate and plot autocorrelation function (ACF)
max_lag = 149  # Consider lags up to half of the length of the time series

autocorrelation = np.zeros(max_lag)
for shift in range(1, max_lag):
    correlation = np.corrcoef(time_series[:-shift], time_series[shift:])[0,1]
    autocorrelation[shift]=correlation

plt.stem(autocorrelation)
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation Coefficient')
plt.show()
print(autocorrelation[])
```
